```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
features=['age','gender','Tot. Bilirubin lvl.','Direct Bilirubin','Alkphos Alkaline Phosphotase','Sgpt Alamine Aminotransferase','Sgot Aspartate Aminotransferase','Total Protiens','Albumi','Albumin-Globulin Ratio','Selector']
data=pd.read_csv('/home/aritra/Downloads/Indian Liver Patient Dataset (ILPD).csv',names=features)
```

```python
from sklearn.preprocessing import Imputer
```

```python
putter=Imputer(strategy='mean',missing_values='NaN')
putter.fit(data[['Albumin-Globulin Ratio']])
data[['Albumin-Globulin Ratio']]=putter.transform(data[['Albumin-Globulin Ratio']])
```

```python
data.info()
```

AFTER HAVING FILLED ALL THE MISSING VALUES

```python
df=pd.get_dummies(data[['gender']])
df.head()
```

```python
data_final=pd.concat([df,data],axis=1)
```

```python
data_final.head()
```

```python
data_final=data_final.drop(['gender'],axis=1)
```

```python
data_final.head()
```

```python
data_final['Selector']=data_final['Selector'].map({2:0,1:1})
```

```python
data_final.head()
```

```python
plt.subplots(figsize=(30,15))
plt.title('CORRELATION MAP')
sns.heatmap(data_final.corr(),linewidths=0.25, vmax=1.0, square=True,annot=True)
plt.show()
```

```python
x=data_final[['age','Tot. Bilirubin lvl.','Direct Bilirubin','Alkphos Alkaline Phosphotase','Sgpt Alamine Aminotransferase','Sgot Aspartate Aminotransferase']]
```

```python
x.head()
```

```python
y=data_final['Selector']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,test_size=0.2,random_state=42)
```

```python
mms=MinMaxScaler()
x_train=mms.fit_transform(x_train)
x_test=mms.transform(x_test)
```

```python
x_test
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
```

```python
classifier=RandomForestClassifier(n_estimators=50,random_state=50)
classifier.fit(x_train,y_train)
print("Accuracy:", classifier.score(x_test, y_test))
print("Precision:", precision_score(y_test, classifier.predict(x_test)))
print("Recall:", recall_score(y_test, classifier.predict(x_test)))
print("F-1 score:", f1_score(y_test, classifier.predict(x_test)))
```

```python
y_predict=classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
results=confusion_matrix(y_test,y_predict)
```

```python
print(results)
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
y_pred=nb.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
result=confusion_matrix(y_test,y_pred)
print(result)
```

```python
print("Accuracy:", nb.score(x_test, y_test))
print("Precision:", precision_score(y_test, nb.predict(x_test)))
print("Recall:", recall_score(y_test,nb.predict(x_test)))
print("F-1 score:", f1_score(y_test,nb.predict(x_test)))
```
