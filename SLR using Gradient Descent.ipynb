```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.array([92,56,88,70,80,49,65,35,66,67])
y=np.array([98,68,81,80,83,52,66,30,68,73])
#x=np.array([1,2,3,4,5])
#y=np.array([5,7,9,11,13])
```

```python
def gradientDescent(x,y):
    mCurrent=bCurrent=0
    iteration=1000
    learningRate=0.000001
    n=len(x)
    
    for i in range(iteration):
        y_predict=mCurrent*x+bCurrent
        mse=(1/n)*np.sum([val**2 for val in (y-y_predict)])
        mPartDerivative=-(2/n)*np.sum(x*(y-y_predict))
        bPartDerivative=-(2/n)*np.sum(y-y_predict)
        mCurrent=mCurrent-(learningRate*mPartDerivative)
        bCurrent=bCurrent-(learningRate*bPartDerivative)
        print('m is {} and b is{} and iter is {} mse is  {}'.format(mCurrent,bCurrent,i,mse))
    
    return(mCurrent,bCurrent)
    
```

```python
def performLinearRegression(x,y):
    temp=gradientDescent(x,y)
    m=round(temp[0],3)
    b=round(temp[1],3)
    y_predictFinal=m*x+b
    makeplots(x,y,y_predictFinal)
    
```

```python
def makeplots(x,y,y_predictFinal):
    plt.figure(1)
    plt.scatter(x,y,color='m',s=60)
    plt.plot(x,y_predictFinal,color='g',linewidth=2)
    plt.xlabel('x')
    plt.ylabel('y')
    print(y_predictFinal)
    
    plt.figure(2)
    c=[val for val in np.linspace(30,100,10)]
    plt.plot(c,y,color='b')
    plt.plot(c,y_predictFinal,color='r')
    plt.title('Actual vs Predict')
    plt.legend(['actual','predict'])
    
    plt.figure(3)    
    plt.plot(c,(y-(y_predictFinal)),color='r',linewidth=3)
    plt.title('Errors')
    
    plt.show()
    
    from sklearn.metrics import r2_score
    rValue=r2_score(y,y_predictFinal)
    print('The rvalue of the model is:'+str(rValue))
```

```python
def main():
    performLinearRegression(x,y)
```

```python
if __name__=='__main__':
    main()
```
